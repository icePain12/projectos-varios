```python
Ci = float(input("Capital Inicial: "))
nt = int(input("Periodos: "))
r = int(input("Tasa de Interes por periodo: "))
pago_capital = Ci/r
periodo_actual = 1
pago_intereses = Cf / nt
pago_total = pago_capital + pago_intereses
nt_actual = nt

while periodo_actual <= nt:
    pago_capital = Ci/r
    Cf = Ci * ((1 + (r/100)/nt_actual)**nt_actual)
    pago_intereses = Cf / nt_actual
    pago_total = pago_capital + pago_intereses
    print(f'Para el periodo: {periodo_actual}')
    print(f'El Capital Final es: ${round(Cf,2)}')
    print(f'Pago Mínimo (Intereses): ${round(pago_intereses,2)}')
    print(f'Pago a Capital: ${round(pago_capital,2)}')
    print(f'Pago para no generar intereses: ${round(pago_total,2)}')
    Ci=Cf-pago_intereses
    print(f'El Capital Final para el proximo periodo es: ${round(Ci,2)} ')
    print("\n")
    nt_actual=nt_actual-1
    print(nt_actual)
    print("\n")
    periodo_actual
```
